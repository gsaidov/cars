```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression, Lasso
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.head(4)
```

```python
pd.plotting.scatter_matrix(df, figsize = (10, 12))
```

```python
y = df['mpg']
```

```python
X = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model', 'origin']]
```

```python
X.head()
```

```python
X = X[['displacement', 'horsepower', 'weight']]
```

```python
X.shape
#y.shape
```

```python
reg = LinearRegression().fit(X[['displacement', 'horsepower', 'weight']], y)
reg.score(X, y)
```
